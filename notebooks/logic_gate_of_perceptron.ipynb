ANDをパーセプトロンで作る

```python
A = [0, 1, 0, 1]
B = [0, 0, 1, 1]

def logic_and(x1, x2):
  if x1 == 1 and x2 ==1:
    return 1
  else:
    return 0

def disp_results(func):
  print(f"A, B = Z")
  print(f"--------")
  for i in range(4):
    Z = func(A[i], B[i])
    print(f"{A[i]}, {B[i]} = {Z}")

disp_results(logic_and)
```

```python
def perceptron_add(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.8
  ans = w1 * x1 + w2 * x2
  if ans > theta:
    return 1
  else:
    return 0

disp_results(perceptron_add)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fillscolors(data):
  return "#ffc2c2" if data > 0 else "#c6dced"
def dotscolors(data):
  return "#ff0e0e" if data > 0 else "#1f7764"

def plot_perceptron(func, x1, x2):
  plt.figure(figsize=(6, 6))
  xx, yy = np.meshgrid(
      np.linspace(-0.25, 1.25, 200),
      np.linspace(-0.25, 1.25, 200))
  xx = np.array(xx).flatten()
  yy = np.array(yy).flatten()
  fills = []
  colors = []
  for i in range(len(xx)):
    fills.append(func(xx[i], yy[i]))
    colors.append(fillscolors(fills[i]))
  plt.scatter(xx, yy, c=colors)

  dots = []
  colors = []
  for i in range(len(x1)):
    dots.append(func(x1[i], x2[i]))
    colors.append(dotscolors(dots[i]))

  plt.scatter(x1, x2, c=colors)
  plt.xlabel("A")
  plt.ylabel("B")
  plt.show()

plot_perceptron(perceptron_add, A, B)
```

ORをパーセプトロンで作る

```python
def perceptron_or(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.2
  ans = w1 * x1 + w2 * x2
  if ans > theta:
    return 1
  else:
    return 0

disp_results(perceptron_or)
```

```python
plot_perceptron(perceptron_or, A, B)
```

NANDをパーセプトロンで作る

```python
def perceptron_nand(x1, x2):
  w1, w2, theta = -0.5, -0.5, -0.9
  ans = w1 * x1 + w2 * x2
  if ans > theta:
    return 1
  else:
    return 0

disp_results(perceptron_nand)
```

```python
plot_perceptron(perceptron_nand, A, B)
```

XORをパーセプトロンで作る

```python
def perceptron_xor(x1, x2):
  if perceptron_or(x1, x2) > 0:
    s1 = 1
  else:
    s1 = 0

  if perceptron_nand(x1, x2) > 0:
    s2 = 1
  else:
    s2 = 0

  ans = perceptron_add(s1, s2)

  if ans > 0:
    return 1
  else:
    return 0

disp_results(perceptron_xor)
```

```python
plot_perceptron(perceptron_xor, A, B)
```
